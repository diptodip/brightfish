# **Environments**

We showcase various environments that can be used for testing a zebrafish model.

# Imports

```python
import torch
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import sys
sys.path.append('../') # assumes notebook started from brightfish/notebooks
import brightfish
import brightfish.environment as environment
```

# Sinusoidal Environment

We create a sinusoidal line environment and then plot the initial status (with normalized brightness on the y-axis and position on the x-axis):

```python
e = environment.SinusoidalLine(201)
fig = plt.figure()
plt.plot(np.arange(201), e.stage.detach().cpu().numpy())
plt.show()
```

We now run the environment for 100 steps and collect the result at each time step:

```python
lines = e.run(100)
```

```python
fig = plt.figure()
plt.plot(np.arange(201), lines[99].detach().cpu().numpy())
plt.show()
```
