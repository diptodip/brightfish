# **Environments**

We showcase various environments that can be used for testing a zebrafish model.

# Imports

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import sys
sys.path.append('../') # assumes notebook started from brightfish/notebooks
import brightfish
import brightfish.environment as environment
```

# Sinusoidal Environment

We create a 2D sinusoidal gradient environment and then plot the initial status (where color shows intensity):

```python
e = environment.SinusoidalGradient((201, 201), static=True)
fig = plt.figure(figsize=(7, 7))
plt.imshow(e.stage, cmap='magma')
plt.colorbar()
plt.clim(0, 1)
plt.title("static sinusoidal environment")
plt.show()
```

We can also make a dynamic sinusoidal environment that moves its peak over time (an oscillating sine wave):

```python
e = environment.SinusoidalGradient((201, 201), static=False)
for i in range(256):
    e.step()
fig = plt.figure(figsize=(7, 7))
plt.imshow(e.stage, cmap='magma')
plt.colorbar()
plt.clim(0, 1)
plt.title("dynamic sinusoidal environment (t=256)")
plt.show()
```
