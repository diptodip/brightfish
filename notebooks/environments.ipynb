# Imports

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import sys
sys.path.append('../') # assumes notebook started from brightfish/notebooks
import brightfish
import brightfish.environment as environment
```

# Sinusoidal Environment

We create a sinusoidal line environment and then plot the initial status (with normalized brightness on the y-axis and position on the x-axis):

```python
e = environment.SinusoidalCircle(201, static=False)
fig = plt.figure()
plt.plot(np.linspace(e.start, e.stop, num=e.shape), e.stage)
plt.xlim(2*np.pi, 0)
plt.title("sinusoidal environment")
plt.xlabel(r"heading ($radians$)")
plt.ylabel("intensity")
plt.show()
```
